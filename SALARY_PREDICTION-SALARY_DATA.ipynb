```python
import pandas as pd
data= pd.read_csv("D:\\MANIPAL_MSIS\\ML_PROJECTS\\DATA\\Salary_Data.csv")
dataset=data
dataset.head()
```

```python
dataset
x=dataset.iloc[:, :1].values
x[:10]
```

```python
dataset
y=dataset.iloc[:, 1:].values
y[:10]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3.0,random_state=0)
x_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color='red')
plt.scatter(x_test,y_test,color='green')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.show()
```

```python
experiance=np.array([[10]])
predicted_salary=regressor.predict(experiance)
print(f"the predicted salary:{predicted_salary}")
```

```python
import pandas as pd
import statsmodels.api as sm

# Corrected dataset
data = {
    'Hours_studied': [2, 4, 6, 8, 10],
    'Classes_Attended': [3, 4, 5, 6, 7],
    'Final_Grade': [60, 70, 85, 90, 100]
}

# Create DataFrame
df = pd.DataFrame(data)

# Define independent variables (add constant for intercept)
X = df[['Hours_studied', 'Classes_Attended']]
X = sm.add_constant(X)

# Define dependent variable
y = df['Final_Grade']

# Build OLS model
model = sm.OLS(y, X).fit()

# Predict for new input: 7 hrs studied, 5 classes attended
new_data = pd.DataFrame({'const': [1], 'Hours_studied': [7], 'Classes_Attended': [5]})
prediction = model.predict(new_data)

print("Predicted Final Grade:", prediction.iloc[0])
```
